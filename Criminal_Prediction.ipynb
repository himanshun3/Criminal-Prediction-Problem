```python
import pandas as pd
import numpy as np
df_train=pd.read_csv('criminal_train.csv')
criminal=df_train['Criminal']
df_test=pd.read_csv('criminal_test.csv')
data=pd.concat([df_train.drop(['Criminal'],axis=1), df_test])
data=data[['PERID','IIHHSIZ2','IIKI17_2','IIHH65_2','PRXRETRY','HLCNOTMO','IRCHMPUS','CELLNOTCL','CELLWRKNG','IIFAMSOC','IRFAMSSI','IRFSTAMP','IIFSTAMP','IRFAMPMT','IRFAMSVC','IRWELMOS','IIWELMOS','IIPINC3','GOVTPROG','TOOLONG','TROUBUND','MAIIN102','AIIND102','ANALWT_C','VESTR','VEREP']]
data_train = data.iloc[:45718]
data_test= data.iloc[45718:]
X=data_train.values
y=criminal.values
test=data_test.values
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X,y)
y_pred=clf.predict(test)
df_test['Criminal']=y_pred
df_test[['PERID', 'Criminal']].to_csv('prediction.csv', index=False)
```
